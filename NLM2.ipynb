Calculating probabilities

```python

import pickle
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
with open("english_unigrams_pickle.pkl", 'rb') as file:
  english_unigrams = pickle.load(file)
with open("english_bigrams_pickle.pkl", 'rb') as file:
  english_bigrams = pickle.load(file)
with open("french_unigrams_pickle.pkl", 'rb') as file:
  french_unigrams = pickle.load(file)
with open("french_bigrams_pickle.pkl", 'rb') as file:
  french_bigrams = pickle.load(file)
with open("italian_unigrams_pickle.pkl", 'rb') as file:
  italian_unigrams = pickle.load(file)
with open("italian_bigrams_pickle.pkl", 'rb') as file:
  italian_bigrams = pickle.load(file)

v = len(english_unigrams) + len(french_unigrams) + len(italian_unigrams)
def prob(text, unigram_dict, bigram_dict, v):
  unigrams_test = word_tokenize(text)
  bigrams_test = list(ngrams(unigrams_test, 2))
  p_laplace = 1
  for bigram in bigrams_test:
    b = bigram_dict[bigram] if bigram in bigram_dict else 0
    u = unigram_dict[bigram[0]] if bigram[0] in unigram_dict else 0
    p_laplace = p_laplace * (b + 1) / (u + v)

  print(p_laplace)
  return p_laplace

with open("test.txt", 'r', encoding='utf=8') as file:
  text = file.readlines()
with open("predicted_languages.txt", "w") as file:
    for t in text:
      english = prob(t, english_unigrams, english_bigrams, v)
      french = prob(t, french_unigrams, french_bigrams, v)
      italian = prob(t, italian_unigrams, italian_bigrams, v)

      if english and french and italian:
        max_language_prob = max(english, french, italian)
        if max_language_prob == english:
          file.write("English\n")
        elif max_language_prob == french:
          file.write("French\n")
        elif max_language_prob == italian:
          file.write("Italian\n")
      else:
        file.write("Unknown\n")


```

Output my accuracy

```python
with open("sol.txt", 'r', encoding='utf-8') as file:
  sol = file.readlines()

sol = ["".join(x for x in label if not x.isdigit() and x != ' ') for label in sol]

with open("predicted_languages.txt", 'r', encoding='utf-8') as file:
  predict = file.readlines()

correct_predictions = sum(1 for s, p in zip(sol, predict) if s == p)
total_predictions = len(sol)
accuracy = correct_predictions / total_predictions

print("Accuracy:", accuracy)





```
