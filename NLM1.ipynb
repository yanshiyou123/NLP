Build separate language models for 3 languages

```python
import os
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
def slp(filename):
  with open(filename, 'r', encoding='utf=8') as file:
    context = file.read().replace('\n', ' ')

  tokens = word_tokenize(context)

  unigrams = word_tokenize(context)
  bigrams = list(ngrams(unigrams, 2))

  udict = {u:unigrams.count(u) for u in set(unigrams)}
  bdict = {b:bigrams.count(b) for b in set(bigrams)}

  return udict, bdict
```

Call the function 3 times, once for each training file (you can hard code the file names), pickle the 6 dictionaries, and save to files with appropriate names.

```python
import pickle
english_unigrams_pickle, english_bigrams_pickle = slp("english.txt")
print(english_unigrams_pickle)
print(english_bigrams_pickle)
french_unigrams_pickle, french_bigrams_pickle = slp("french.txt")
print(french_unigrams_pickle)
print(french_bigrams_pickle)
italian_unigrams_pickle, italian_bigrams_pickle = slp("italian.txt")
print(italian_unigrams_pickle)
print(italian_bigrams_pickle)

with open("english_unigrams_pickle.pkl", 'wb') as file:
  pickle.dump(english_unigrams_pickle, file)
with open("english_bigrams_pickle.pkl", 'wb') as file:
  pickle.dump(english_bigrams_pickle, file)
with open("french_unigrams_pickle.pkl", 'wb') as file:
  pickle.dump(french_unigrams_pickle, file)
with open("french_bigrams_pickle.pkl", 'wb') as file:
  pickle.dump(french_bigrams_pickle, file)
with open("italian_unigrams_pickle.pkl", 'wb') as file:
  pickle.dump(italian_unigrams_pickle, file)
with open("italian_bigrams_pickle.pkl", 'wb') as file:
  pickle.dump(italian_bigrams_pickle, file)
```
